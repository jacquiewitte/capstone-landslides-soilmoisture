#### Change directory to where all the Capstone data reside

```python
import os
from glob import glob

import earthpy as et
import xarray as xr
```

#### Directory where GPM data reside

```python
esa_dir = os.path.join(et.io.HOME,
                        'earth-analytics',
                        'data', 'capstone', 
                        'esa_soil_moisture')
# Find the files, make them sorted
esa_files = sorted(glob(os.path.join(esa_dir, '*nc')))
len(esa_files)
```

#### Read the data into a list

```python
esa_files2006_2009=esa_files[0:1095]
```

```python
esa_files2006_2011=esa_files[0:1825]
esa_files2012_2017=esa_files[1826:len(esa_files)]
```

```python
esa_all = []
for i, file in enumerate(esa_files2006_2009):
    print(file)
    data = xr.open_dataset(file).squeeze()
    esa_all.append(data)
```

#### Turn list of data into a single xarray object. This takes a very long time! 

```python
esa_xr = xr.concat(esa_all, dim="time")
```

#### Save dataArray to a netCDF4 file

```python
esa_xr.to_netcdf(os.path.join(esa_dir, 'esa_soilmoisture_2006-2009_saved_on_disk.nc'))
```
